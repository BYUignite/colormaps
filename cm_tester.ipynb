```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
from byuignite_jet_cm import *
```

```python
data = np.loadtxt('jet.pgm')
d = np.reshape(data, (469,792))
y = np.arange(469)
x = np.arange(792)
X,Y = np.meshgrid(x,y)

fig, ax = plt.subplots(1,1)
im = ax.contourf(X,Y,d,255, cmap=byuignite_jet_r)
ax.set_aspect('equal')
fig.colorbar(im, ax=ax, orientation='horizontal');
```
