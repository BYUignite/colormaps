```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

from byuignite_jet_cm import *
from byuignite_inferno_white_cm import *
from byuignite_viridis_white_cm import *
```

```python
img = Image.open('mona_lisa.pgm')
IMG = np.asarray(img)
IMG = np.flip(IMG, axis=0)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
con = ax.contourf(IMG, cmap='jet')#, levels=50)
ax.set_aspect('equal')
fig.colorbar(con, ax=ax);
ax.set_xticks([]); ax.set_yticks([]);
```

```python
fig, ax = plt.subplots(figsize=(10,10))
con = ax.contourf(IMG, cmap=byuignite_jet_r)#, levels=50)
ax.set_aspect('equal')
fig.colorbar(con, ax=ax);
ax.set_xticks([]); ax.set_yticks([]);
```
